# Data Preprocessing

Converting categorical values into numerical values

As there are many columns taking values more than 2(yes or no) thus we need to use Label Encoder to convert them into strings

```python
from sklearn.preprocessing import LabelEncoder
Le = LabelEncoder()

for i in data.columns:
    if data[i].dtypes=="object":
        data[i]=Le.fit_transform(data[i].values.reshape(-1,1))
```

```python
#let's split our data into features and label
x=data.drop("Churn",axis=1)
y=data['Churn']
```

# Scaling the data using MinMax Scaler

```python
from sklearn.preprocessing import MinMaxScaler
mns=MinMaxScaler()
sclaer=mns.fit_transform(x)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier 
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics #accuracy measure
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.metrics import plot_roc_curve
```

```python
x_train, x_test, y_train, y_test = train_test_split(sclaer,y, test_size=0.3, random_state=49)
```

# Model Building

```python
lr=LogisticRegression()
knn=KNeighborsClassifier()
dt=DecisionTreeClassifier()
rf=RandomForestClassifier()
```

```python
#Training of Models starts here

lr.fit(x_train,y_train)
knn.fit(x_train,y_train)
dt.fit(x_train,y_train)
rf.fit(x_train,y_train)

print('Training completed')
```

```python
#Let's check the scores now

print(lr.score(x_test,y_test))
print(knn.score(x_test,y_test))
print(dt.score(x_test,y_test))
print(rf.score(x_test,y_test))
```

```python
#Confusion Matrix

show=plot_roc_curve(dt,x_train,y_train)
plot_roc_curve(lr,x_train,y_train,ax=show.ax_)
plot_roc_curve(knn,x_train,y_train,ax=show.ax_)
plot_roc_curve(rf,x_train,y_train,ax=show.ax_)
plt.legend(prop={'size':10}, loc='upper right')
plt.show()
```

```python
show=plot_roc_curve(dt,x_test,y_test)
plot_roc_curve(lr,x_test,y_test,ax=show.ax_)
plot_roc_curve(knn,x_test,y_test,ax=show.ax_)
plot_roc_curve(rf,x_test,y_test,ax=show.ax_)
plt.legend(prop={'size':10}, loc='lower right')
plt.show()
```

```python
lr_pred = lr.predict(x_test)
knn_pred = knn.predict(x_test)
dt_pred = dt.predict(x_test)
rf_pred = rf.predict(x_test)
```

```python
print('Logistic Regression:',accuracy_score(y_test,lr_pred))
print('KNN :',accuracy_score(y_test,knn_pred))
print('Decision Tree :', accuracy_score(y_test,dt_pred))
print('Random Forest :',accuracy_score(y_test,rf_pred))
```

From the aboove results, it is clear that Logistic Regression model is working best for this data, soo let us proceed doing Hyperparameter tuning on LogisticRegression()

```python
# Let's do Hyperparameter tuning using GridSearchCV
grid={"C":np.logspace(-3,3,7), "penalty":["l1","l2"]}
```

```python
lr_cv=GridSearchCV(lr,grid,cv=10)
```

```python
lr_cv.fit(x_train,y_train)

print("tuned hpyerparameters :(best parameters) ",lr_cv.best_params_)
print("accuracy :",lr_cv.best_score_)
```

Thus our predictions are almost 80% accurate with Logistic Regression, i.e. we have identified 80% of the churn analysis prediction correctly for our Logistic Regression with grid search.
